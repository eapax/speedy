```python
import matplotlib.pyplot as plt
```

```python


    


def potential(r):
    
    l2 = r / (1-3.0/r)
    pot = -1/r * (1+l2/r**2)
    
    return l2/(2*r**2) +pot
    
    
    
    
```

```python
import numpy as np
r = np.linspace(1.1,10,100)
y = potential(r)
```

```python
plt.figure(figsize=(15,15))
plt.plot(r,y)
plt.ylim(-2,2)
plt.axvline(3, c='r', linestyle='--')
plt.axvline(6, c='r', linestyle='--')
```
