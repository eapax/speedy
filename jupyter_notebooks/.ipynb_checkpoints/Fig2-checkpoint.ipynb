```python
#Import everything
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import time
import glob
import pandas as pd
import seaborn as sns
import sys
from collections import defaultdict
from tqdm.notebook import tqdm
import matplotlib as mpl






def extract_file_info(d):
    
    #Mapping between levels
    level_dict = {'L1': 'El Nino Static SST. ablco2=6',
                  'L2': 'EC Earth SST. ablco2=6',
                  'L3': 'El Nino Static SST. ablco2=21',
                  'L4': 'EC Earth SST. ablco2=21'
                  }

    #Create some useful labels
    label = d.split('/')[-1]
    level = label.split('_')[1]
    prec = label.split('_')[2]
    rounding = label.split('_')[3]
    title = level_dict[level]
    
    keys = ['label', 'level', 'precision','title']
    values = [label,level,prec+'_'+rounding,title]
    dictionary = {keys[i]: values[i] for i in range(len(keys))}
    
    return dictionary




def surface_slice(y,weights):
    
 
    #Get values at surface
    y_surface =  y[:,0]   
    
    ygrid = y_surface.weighted(weights).mean(("forecast_period"))# average over time
    
    
   # display(y_surface)
    #display(y_surface.weighted(weights))
#     display(ygrid)
#     display(y_surface.mean(("forecast_period")))
#     display(weights)
    
    #sys.exit()
    
    return ygrid.to_dataframe()


def extract_file_data(df,columns,weights):
    
    
    data = [surface_slice(getattr(df,
                                  columns[k]),
                                   weights = weights) for k in range(len(columns))][0]
    
    
    return data #This is a pandas df

    
    
def process_nc_file(fname,weights,directory,true_lat):    
    
        #Get data 
        df = xr.open_dataset(fname)
        
        
        #Reset the latitude
        df = df.assign_coords(latitude=(true_lat))
        

       
        #Get the data you want     
        columns_3d  = ['temperature']
        data3d = extract_file_data(df,columns_3d,weights)

                
        #Get meta information
        f_info = extract_file_info(directory)
        index = data3d.index
        df_info = pd.DataFrame(f_info,index=index)
      
        #Create pandas df and append
        df_meta = pd.concat([data3d,df_info,],axis=1) 
        
        return df_meta
        
    
        
def process_all_data(all_directories,weights,true_lat):

    
    #Empty arrays to hold data
    dfs = []
    

    for d in tqdm(all_directories): #for every directory
        print ('-----------',d)
        counter = 1
        df_LI = [] # df of solution level i
        nc_files = sorted(glob.glob(d+'/model_output*.nc'))
                
        for n in tqdm(nc_files): #for every model.nc file
            print (n, counter)

            df_meta = process_nc_file(n,weights,d,true_lat)
            df_meta['ID'] = counter #index to show which time period is is from
            print('ID = ', counter)
            df_LI.append(df_meta)
            counter = counter + 1
            
            
        #df_LI = pd.concat(df_LI,ignore_index=True)
        df_LI = pd.concat(df_LI)
        #df_LI['X'] = np.arange(len(df_LI))
        
        #Add append to bigger array
        dfs.append(df_LI)
            
    #Bring all together
    df = pd.concat(dfs)
    
    return df #.reset_index()

    
def get_global_weights(root):
    
     #Get the latitude weights from a special location
    f = root + 'speedyoneFIG1_L2_52_RN_10year/model_output00001.nc'
    df = xr.open_dataset(f)
    
    temp_file = df.temperature
    weights = np.cos(np.deg2rad(temp_file.latitude))
         
    weights.name = "weights"
    
    
    return weights, temp_file.latitude
    

def process_node(node):
    
    root= '/network/group/aopp/predict/TIP016_PAXTON_RPSPEEDY/speedyone/'+node
    
   
    #Get the global weights
    weights,true_lat = get_global_weights(root)
    
    
    
    #iterate over every directory
    all_dirs = glob.glob(root+'speedyone*')     
    df = process_all_data(all_dirs,weights,true_lat)
    
     
    
    return df.reset_index()#.set_index(['latitude', 'longitude', 'ID'])





```

```python
node = 'paper/Fig1_10year/'

df = process_node(node)
```

```python
df
```

```python
#We need to weight the temperature by the latitude
df['temperature_weighted'] = df.temperature*np.cos(np.deg2rad(df.latitude))
df = df.set_index(['latitude', 'longitude', 'ID'])
display(df)

```

Before we split into double precision and "the rest".

Now instead we are just going to grab the 10RN and 10SR solutions:

```python
df_weighted = df.copy() #Make a copy, used later
dfRN = df.query("precision == '10_RN'")
dfSR = df.query("precision == '10_SR'")
```

```python
#Pivot data to allow us to subtract to get difference
df_pivot = pd.merge(dfRN, dfSR['temperature_weighted'], left_index=True,right_index=True,suffixes=(None,"_SR"))
df_pivot['delta'] = df_pivot['temperature_weighted'] - df_pivot['temperature_weighted_SR'] 


display(df_pivot)


```

```python
#And now average over the total time (set by ID)

df_pivot_averaged = df_pivot.groupby(['latitude', 'longitude','precision']).mean()
display(df_pivot_averaged)

```

```python
def plot_stack(df,save):   
    
    fig = plt.figure(figsize=[24,24])
    ax1 = fig.add_subplot(211, projection=ccrs.PlateCarree(central_longitude=0)) #RN 
    ax2 = fig.add_subplot(212, projection=ccrs.PlateCarree(central_longitude=0)) #SR


    dataRN = df.query('precision == "10_RN"').to_xarray()
    dataSR = df.query('precision == "10_SR"').to_xarray()

    
    cmap = 'bwr'
    qRN = dataRN['delta'].plot(ax=ax1, 
                 cmap=cmap,
                 transform=ccrs.PlateCarree(),
                 add_colorbar=False)
    qSR = dataSR['delta'].plot(ax=ax2, 
                 cmap=cmap,
                 transform=ccrs.PlateCarree(),
                 add_colorbar=False)


    ax1.coastlines()
    ax2.coastlines()

    ax1.set_title(r'Round to Nearest',fontsize = 20 )
    ax2.set_title(r'Stochastic Rounding',fontsize = 20 )



    vlow = -1
    vhigh= 1
    qRN.set_clim(vmin=vlow, vmax=vhigh)
    qSR.set_clim(vmin=vlow, vmax=vhigh)



    fig.subplots_adjust(right=0.8,wspace = 0.1,hspace=0.0)
    cbar_ax = fig.add_axes([0.85, 0.15, 0.02, 0.7])
    cbar = fig.colorbar(qRN, cax=cbar_ax)

    cbar.ax.set_ylabel(r'$\delta$ K', fontsize = 25, rotation = 0 )


    #cbar.ax.set_ylabel('K', fontsize = 20, rotation = 0 )
    cbar.ax.tick_params(labelsize=20)
    
    if save:
        plt.savefig('figs/Fig2.png', dpi=300, bbox_inches='tight')
    
    
    
    

    plt.show()

```

```python
save = True
plot_stack(df_pivot_averaged,save)
```
